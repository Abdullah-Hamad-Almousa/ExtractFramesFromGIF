```python
from PIL import Image
import os

def extract_gif_frames(gif_path, output_folder):
    gif = Image.open(gif_path)

    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    frame_number = 1  # Start from 1 instead of 0
    while True:
        try:
            gif.save(
                f"{output_folder}/image_{frame_number}.png",  # Changed format: image_1.png, image_2.png, etc.
                "PNG"
            )
            frame_number += 1
            gif.seek(frame_number)
        except EOFError:
            break

    print(f"Extracted {frame_number - 1} frames to {output_folder}")  # Adjust the count since we started from 1

# âœ… Since image folder is in same root, use:
extract_gif_frames("image/background.gif", "extracted_frames")
```
